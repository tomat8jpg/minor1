```python
import pandas as pd
```

```python
df = pd.read_csv('Desktop/что-то нужное или нет/APXNB/прога/ufc-master.csv', low_memory=False)
df.head()
```

```python
from ludwig.api import LudwigModel
```

```python
model_definition = {
    'input_features':[
        {'name': 'total_title_bout_dif', 'type': 'numerical'},
        {'name': 'B_current_lose_streak', 'type': 'numerical'},
        {'name': 'R_current_lose_streak', 'type': 'numerical'},
        {'name': 'R_losses', 'type': 'numerical'},
        {'name': 'R_losses', 'type': 'numerical'}
    ],
    'output_features':[
        {'name': 'Winner', 'type': 'category'}
    ]
}
```

```python
model = LudwigModel (model_definition)
train_stats = model.train(df)
```

```python
train_stats
```

```python
from ludwig.visualize import *
```

```python
learning_curves(train_stats, 'Winner')
```
